  **Establishing Relationships Between Tables**

Before we can merge the tables into a single analytical dataset, we need to verify that the relational keys (such as `subject_id` and `hadm_id`) are consistent across tables.  
In this step, we test the relationships between the `patients`, `admissions`, `diagnoses_icd`, and `procedures_icd` tables to confirm that each admission and diagnosis can be correctly linked to the corresponding patient.


```python
# Step 3: Establishing Relationships

# Patients ↔ Admissions (subject_id)
patients_in_adm = admissions_df['subject_id'].isin(patients_df['subject_id']).mean()
print(f"Admissions linked to patients: {patients_in_adm*100:.2f}%")

# Admissions ↔ Diagnoses (hadm_id)
adms_in_diag = diagnoses_df['hadm_id'].isin(admissions_df['hadm_id']).mean()
print(f"Diagnoses linked to admissions: {adms_in_diag*100:.2f}%")

# Admissions ↔ Procedures (hadm_id)
adms_in_proc = procedures_df['hadm_id'].isin(admissions_df['hadm_id']).mean()
print(f"Procedures linked to admissions: {adms_in_proc*100:.2f}%")

# Quick counts of unique IDs across tables
print("\nUnique IDs across tables:")
print(f"Patients: {patients_df['subject_id'].nunique()}")
print(f"Admissions: {admissions_df['hadm_id'].nunique()}")
print(f"Diagnoses: {diagnoses_df['hadm_id'].nunique()}")
print(f"Procedures: {procedures_df['hadm_id'].nunique()}")
```

**Step – Data Cleaning and Preprocessing**

Once the relationships between the tables have been confirmed, we perform preliminary data cleaning to ensure consistency across all tables.  
This involves converting date columns to proper `datetime` formats, checking for missing values, standardizing column names (e.g., converting to lowercase), and removing any duplicate records.  
These preprocessing steps are essential to avoid errors during feature engineering and model training.


```python
# Step 4: Preprocessing for Consistency

# Convert datetime columns in Admissions
date_cols = ['admittime', 'dischtime', 'deathtime']
for col in date_cols:
    if col in admissions_df.columns:
        admissions_df[col] = pd.to_datetime(admissions_df[col], errors='coerce')

# Check for missing values in each table
print("Missing values summary:")
print("\nPatients:")
print(patients_df.isnull().sum())
print("\nAdmissions:")
print(admissions_df.isnull().sum())
print("\nDiagnoses:")
print(diagnoses_df.isnull().sum())
print("\nProcedures:")
print(procedures_df.isnull().sum())

# Standardize column names to lowercase for easier merging
patients_df.columns = patients_df.columns.str.lower()
admissions_df.columns = admissions_df.columns.str.lower()
diagnoses_df.columns = diagnoses_df.columns.str.lower()
procedures_df.columns = procedures_df.columns.str.lower()

# Drop duplicates if any
patients_df = patients_df.drop_duplicates()
admissions_df = admissions_df.drop_duplicates()
diagnoses_df = diagnoses_df.drop_duplicates()
procedures_df = procedures_df.drop_duplicates()

print("\nPreprocessing complete ✅")
```

**Step – Feature Engineering and Merging of Tables**

Now that we have explored the data, we will combine the relevant tables and engineer features that can be used for model training.  
For this project, we focus on variables that are likely to influence healthcare service demand, such as:

- Patient age  
- Gender  
- Admission type  
- Number of diagnoses assigned during an admission  
- Number of procedures performed during an admission  

These features are created by merging the `patients`, `admissions`, `diagnoses_icd`, and `procedures_icd` tables using the relational identifiers.


```python
# -----------------------------------------
# 1) Merge admissions with patients to get age and gender
# -----------------------------------------
tmp = admissions_df.merge(patients_df[['subject_id', 'gender', 'dob']], on='subject_id', how='left')
tmp['dob'] = pd.to_datetime(tmp['dob'], errors='coerce')
tmp['age'] = tmp['admittime'].dt.year - tmp['dob'].dt.year

# -----------------------------------------
# 2) Add number of diagnoses per admission
# -----------------------------------------
diag_counts = diagnoses_df.groupby('hadm_id').size().reset_index(name='num_diagnoses')
tmp = tmp.merge(diag_counts, on='hadm_id', how='left')
tmp['num_diagnoses'] = tmp['num_diagnoses'].fillna(0)

# -----------------------------------------
# 3) Add number of procedures per admission
# -----------------------------------------
proc_counts = procedures_df.groupby('hadm_id').size().reset_index(name='num_procedures')
tmp = tmp.merge(proc_counts, on='hadm_id', how='left')
tmp['num_procedures'] = tmp['num_procedures'].fillna(0)

# -----------------------------------------
# 4) Select relevant columns for modelling
# -----------------------------------------
model_df = tmp[['age', 'gender', 'admission_type', 'num_diagnoses', 'num_procedures']].copy()

# Encode gender (M=0, F=1)
model_df['gender'] = model_df['gender'].map({'M':0, 'F':1})

# Display the first 5 rows of the engineered dataset
model_df.head()
```

**Step – Model Building and evaluation**

In this step, we train a **Random Forest** classification model using the engineered dataset.  
The goal is to predict the healthcare service type (e.g., Emergency vs Elective) based on patient demographics and clinical features.  
The dataset will first be split into training and test sets to allow unbiased evaluation of the model’s performance.


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# -----------------------------------------
# Encode the target variable (admission_type)
# -----------------------------------------
le = LabelEncoder()
model_df['service_type'] = le.fit_transform(model_df['admission_type'])

# Define features (X) and target (y)
X = model_df[['age', 'gender', 'num_diagnoses', 'num_procedures']]
y = model_df['service_type']

# -----------------------------------------
# Train-test split
# -----------------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

# -----------------------------------------
# Build & train Random Forest classifier
# -----------------------------------------
rf_model = RandomForestClassifier(n_estimators=200, random_state=42)
rf_model.fit(X_train, y_train)

# -----------------------------------------
# Evaluate the model
# -----------------------------------------
y_pred = rf_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"Model Accuracy: {accuracy * 100:.2f}%")
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=le.classes_))
```
